```python
# Variables y tipos de datos

# Asignación de valores a variables
nombre = input("Introduce tu nombre: ").lower()  # Solicita al usuario que ingrese su nombre y lo convierte a minúsculas
edad = int(input("Introduce tu edad: ")) # Solicita al usuario que ingrese su edad y convierte el valor a entero

# Solicitar al usuario que ingrese su altura y validar el formato
altura = input("Introduce tu altura en metros: ")

# Verificar que la entrada cumpla con el formato adecuado
if altura.count('.') == 1:
    partes = altura.split('.')
    if len(partes[0]) <= 1 and len(partes[1]) == 2:
        altura = float(altura)
        print("Altura válida:", altura)
    else:
        print(f"La altura '{altura}' no es válida.")
        print("Por favor, ingresa un valor con máximo 1 dígito antes del punto y exactamente 2 dígitos después del punto.")
else:
    print(f"La altura '{altura}' no es válida.")
    print("Por favor, ingresa un valor en el formato correcto (por ejemplo, 1.56)")

# Solicitar al usuario si es estudiante y asignar el valor correspondiente
respuesta = input("¿Eres estudiante? (Sí/No): ").lower()  # Solicita al usuario que ingrese su respuesta y convierte a minúsculas
es_estudiante = respuesta in ["si", "sí", "no", "NO","noo"]# Verifica si la respuesta se encuentra en la lista de respuestas válidas

# La línea anterior utiliza la expresión 'in' para verificar si la respuesta ingresada se encuentra en la lista ["si", "sí", "no", "NO","noo"].
# La expresión 'in' devuelve True si la respuesta se encuentra en la lista y False si no se encuentra.

# Esta lista incluye diferentes variantes de las respuestas afirmativas y negativas, teniendo en cuenta distintas combinaciones de mayúsculas, minúsculas y acentos.
# Por ejemplo, si el usuario ingresa "Sí" o "NOO", la expresión 'in' evaluará a True porque esas respuestas están presentes en la lista.

# La variable 'es_estudiante' se asigna con el valor de True si la respuesta ingresada se encuentra en la lista de respuestas válidas,
# y se asigna False si la respuesta no se encuentra en la lista.

# Imprimir el contenido de las variables
print("Mi nombre es", nombre)
print("Tengo", edad, "años")
print("Mi altura es", altura, "metros")
print("¿Soy estudiante?", es_estudiante)

# Estructuras de control: condicionales

if edad >= 18:  # Si la edad es mayor o igual a 18
    print("Soy mayor de edad")
else:
    print("Soy menor de edad")

# Estructuras de control: bucles

# Bucle while
contador = 0
while contador < 5:
    print("El contador es", contador)
    contador += 1

# Bucle for
frutas = ["manzana", "banana", "cereza"]
for fruta in frutas:
    print("Me gusta comer", fruta)

# Funciones

# Definición de una función
def calcular_area_rectangulo(base, altura):
    area = base * altura
    return area

# Llamada a la función
base_rectangulo = int(input("Introduce la base del rectángulo: "))
altura_rectangulo = int(input("Introduce la altura del rectángulo: "))
area_rectangulo = calcular_area_rectangulo(base_rectangulo, altura_rectangulo)
print("El área del rectángulo es:", area_rectangulo)

# Módulos

# Importar el módulo math
import math

# Utilizar una función del módulo math
raiz_cuadrada = math.sqrt(25)
print("La raíz cuadrada de 25 es:", raiz_cuadrada)


# CAMBIOS REALIZADOS EN EL CODIGO
#se detallan los cambios implementados en el código desde su versión original(TAREA):

#Se ha añadido una validación para asegurarse de que el formato de la altura ingresada por el usuario sea correcto. Ahora, el programa verifica que la entrada cumpla con los siguientes requisitos:

#Debe haber exactamente un punto decimal en la altura.
#El número antes del punto decimal debe tener un máximo de un dígito.
#El número después del punto decimal debe tener exactamente dos dígitos.
#Estas comprobaciones garantizan que la altura ingresada tenga el formato adecuado y cumpla con los criterios establecidos.

#En caso de que la altura ingresada no cumpla con el formato correcto, se muestra un mensaje de error más elegante y claro. El mensaje de error incluye la respuesta de altura ingresada por el usuario, lo que facilita la identificación del valor incorrecto. Además, se proporciona una explicación detallada de las condiciones que la altura válida debe cumplir.

#Se ha añadido una validación del formato de altura ingresada por el usuario. Ahora, se verifica que la entrada cumpla con las siguientes condiciones:

#Debe haber exactamente un punto decimal en la altura.
#La parte antes del punto decimal debe contener máximo un dígito.
#La parte después del punto decimal debe contener exactamente dos dígitos.
#Estas comprobaciones garantizan que la altura ingresada tenga el formato correcto y cumpla con los requisitos establecidos.

#En caso de que la altura ingresada no cumpla con el formato adecuado, se muestra un mensaje de error más elegante y descriptivo. El mensaje de error incluye la respuesta de altura ingresada por el usuario, lo que facilita la identificación del valor incorrecto. Además, se proporciona una explicación clara de las condiciones que la altura válida debe cumplir.

#Estos cambios mejoran la experiencia del usuario al brindar mensajes de error más informativos y comprensibles.
```
